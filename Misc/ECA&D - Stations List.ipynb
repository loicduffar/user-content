# Map of ECA&D Stations

Stations list: https://knmi-ecad-assets-prd.s3.amazonaws.com/download/stations.txt
Climatologic Data: https://cds.climate.copernicus.eu/cdsapp#!/dataset/insitu-gridded-observations-europe?tab=overview

Author: Loïc Duffar https://github.com/loicduffar/user-content/tree/main/Misc

```python
import pandas as pd

df = pd.read_excel('https://github.com/loicduffar/user-content/blob/main/Misc/ECA&D%20-%20Stations%20List.xlsx?raw=true',
             sheet_name='List', skiprows=17)
df.set_index('STAID', inplace= True, drop=True)
display(df)
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Scattermapbox(
                        mode = "markers",
                        lon = df['LON'].round(2),
                        lat = df['LAT'].round(2),
                        marker = dict(size= 4, color= 'red', opacity= 0.5),
                        text = [str(a).strip() + '-' + str(b) for (a, b) in zip(*[df['STANAME'].values, df['CN'].values])],
                                )
               )
fig.update_layout(  margin = dict(t=0, b=0, r=0, l=0),
                    mapbox = dict(
                                style= "open-street-map",    # open-street-map, carto-positron, stamen-terrain, stamen-toner
#                                center= dict(lon= df['X'].mean(), lat= df['Y'].mean() ),
#                                zoom= zoom
                                )
                 )
fig.show()
```
